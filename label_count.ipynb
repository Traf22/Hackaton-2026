```python
import pandas as pd
import ast

df = pd.read_csv("/Users/longquanwen0813/Downloads/all.tsv", sep="\t")
```

```python
df.columns
```

```python
df["timel_labels_list"] = df["timel_labels_list"].apply(ast.literal_eval)
```

```python
df["num_labels"] = df["timel_labels_list"].apply(len)
```

```python
count_by_num = df["num_labels"].value_counts().sort_index()
count_by_num
```

```python
ratio_by_num = count_by_num / count_by_num.sum() * 100
```

```python
stats = pd.DataFrame({
    "num_labels": count_by_num.index,
    "num_images": count_by_num.values,
    "ratio_percent": ratio_by_num.values
})

stats
```
