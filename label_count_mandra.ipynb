```python
import pandas as pd

path = "/Users/longquanwen0813/Downloads/mandragore_nh_global_kw_grouped.csv"

df = pd.read_csv(path, sep=None, engine="python")  # 自动检测分隔符
df.head()
```

```python
df.columns
```

```python
import pandas as pd

# 如果 mots_cles 可能有空值 / 多余空格，这个写法最稳
df["mots_cles_list"] = (
    df["mots_cles"]
    .fillna("")                              # NaN -> ""
    .astype(str)
    .str.split("|")                          # 用 | 切分
    .apply(lambda xs: [x.strip() for x in xs if x.strip()])  # 去空白+去空元素
)

df["num_labels"] = df["mots_cles_list"].apply(len)
```

```python
count_by_num = df["num_labels"].value_counts().sort_index()
ratio_by_num = count_by_num / count_by_num.sum()

stats = pd.DataFrame({
    "num_labels": count_by_num.index,
    "num_images": count_by_num.values,
    "proportion": ratio_by_num.values
})

stats
```

```python
stats["percent"] = (stats["proportion"] * 100).round(2)
stats
```

```python
q1, q2 = df["num_labels"].quantile([1/3, 2/3])
q1, q2
```

```python
def assign_group_range(n):
    if 0 <= n <= 10:
        return "G1_low"
    elif n <= 21:
        return "G2_mid"
    else:
        return "G3_high"

df["label_group"] = df["num_labels"].apply(assign_group_range)
```

```python
df.groupby("label_group")["num_labels"].agg(["min", "max", "count"])
```
